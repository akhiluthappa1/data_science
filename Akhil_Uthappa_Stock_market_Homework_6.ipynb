### Q3. Implement a trading strategy based on your labels for year 2.


```python
#Main Algorithm
money_available = 100
shares = 0
week_count = 1
buy_flag = False
flag = False #False when cashed out; true when invested in the stock

df_weekly = pd.read_csv("/content/AAPL_weekly_return_volatility_1.csv",sep=",") #data with labels
df_detailed = pd.read_csv("/content/AAPL_weekly_return_detailed (1).csv",sep=",") #data with prices
df_weekly.head()
```

```python
def stock_amount (cash,purchse_value):
    stock_bought = round(cash/purchse_value, 2)
    return stock_bought
def buy_out (shares, buy_out_value):
    total_amount = shares * buy_out_value
    total_amount = round (total_amount, 2)
    return total_amount

for i,r in df_weekly.iterrows():
    if r["Year"] == 2020:
        continue
    label = r["new_label"]
    df_week = df_detailed[(df_detailed["Year"] == 2021)&(df_detailed["Week_Number"]==week_count)]
    
    if buy_flag == False:
        if label == "undefined":
            week_count += 1
        elif label == "red":
            week_count += 1
        elif label == "green":
            buying_price = df_week.iloc[0]["Open"]
            shares = stock_amount(money_available, buying_price)
            money_available = 0
            buy_flag = True
            flag = True
            week_count += 1
    else:
        if label == "red":
            if flag == True:
                previous_week = week_count - 1
                df_week = df_detailed[(df_detailed["Year"] == 2021) & (df_detailed["Week_Number"] == previous_week)]
                selling_price = df_week.iloc[-1]["Adj Close"]
                new_cash_amount = buy_out(shares, selling_price)
                money_available = new_cash_amount
                shares = 0
                week_count += 1
                flag = False
            elif flag == False:
                week_count += 1
        if label == "green":
            if flag == True:
                week_count += 1
            elif flag == False:
                buying_price = df_week.iloc[0]["Open"]
                shares = stock_amount(money_available, buying_price)
                money_available = 0
                week_count += 1
                flag = True
    print(week_count, label, flag, shares, money_available)
if shares!=0:
    df_week = df_detailed[(df_detailed["Year"] == 2021)&(df_detailed["Week_Number"]==(week_count-1))]
    selling_price = df_week.iloc[-1]["Adj Close"]
    new_cash_amount = buy_out(shares, selling_price)
    money_available = new_cash_amount
    shares = 0
print(shares, money_available)
```
