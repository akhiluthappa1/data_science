## 11. Compute the rank of countries (by population) in 1970 and compute the difference in rank from 1970 to 2020. Which 5 countries experienced the largest positive change (by value, not percentage) in rank and which 5 countries experiences the largest negative change in rank 

```python
df_1970 = df.sort_values(by=['1970 Population'],ascending=False)
df_1970['rank_population_1970'] = df['1970 Population'].rank(method='max', ascending=False)
df_1970['rank_population_2020'] = df['2020 Population'].rank(method='max', ascending=False)
df_1970['ranking_change'] = df_1970['rank_population_1970']-df_1970['rank_population_2020']
df_1970 = df_1970.sort_values(by=['ranking_change'],ascending=False)
df_1970_tail = df_1970.tail().sort_values(by=['ranking_change'],ascending=True)
print("LARGEST POSITIVE CHANGE---->\n \n", df_1970.head()[['Country', 'ranking_change']])
print("LARGEST NEGATIVE CHANGE---->\n \n", df_1970_tail[['Country', 'ranking_change']])
 
```
