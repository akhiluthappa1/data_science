## 1. load the dataset csv file as a dataframe using Pandas

```python
import pandas as pd
import numpy as np

df = pd.read_csv("./ds_salaries.csv", sep=",")
print(df.to_string()) 
```

## 2. compute average salary by year?

```python
df_2=df.groupby(['work_year'])["salary_in_usd"].mean()
df_2
```

## 3. which year was the lowest and which year was the highest?

```python
print("Minimum\n", df_2.iloc[[0]], "\n\nMaximum\n", df_2.iloc[[2]])
```

## 4. For each experience level, compute the average salary (over 3 years) for each job title?


```python
df_3 = df.groupby(['experience_level', 'job_title', 'salary_in_usd'])["salary_in_usd"].mean()
print(df_3.to_string())
```

## 5. For which job title was it the minimum and maximum?

```python
print("Maximum", df[df.salary_in_usd == df.salary_in_usd.max()])
print("\n Minimum", df[df.salary_in_usd == df.salary_in_usd.min()])
```

## 6. For each year and job title compute the annual salary?

```python
df_4 = df.groupby(['work_year', 'job_title'])["salary_in_usd"].mean()
print(df_4.to_string())
```

## 7. Which job title experienced the maximum change in salary? Which job title had the minimum change.

```python
df_5 = df.groupby('job_title')['salary_in_usd'].apply(lambda x: (x.max()-x.min())).reset_index(name='salary_difference')
df_5 = df_5.sort_values(by=['salary_difference'])
print(df_5)
print("Minimum salary change---->", df_5.iloc[0])
print(len(df_5))
for i in range(len(df_5)):
  if df_5.iloc[i]['salary_difference']!=0:
    print("Minimum salary change which is not 0---->", df_5.iloc[i])
    break
print("Maximum salary change---->", df_5.iloc[-1])
```

## 8. compute the average salaries for each remote ratio? how many entries are there?

```python
df_5 = df.groupby(['remote_ratio']).agg({"salary_in_usd": ['mean', 'count']})
df_5
```

## 9. which company locations pay the highest salary and lowest salaries.

```python
df_6 = df.groupby(['company_location'])['salary_in_usd'].max()
df_6 = df_6.sort_values()
df_6
print("MAX- ", df_6.iloc[[-1]], "\n MIN-", df_6.iloc[[0]])
```
