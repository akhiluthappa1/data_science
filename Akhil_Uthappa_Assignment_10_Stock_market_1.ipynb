```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.width",1000)
df = pd.read_csv("/content/AAPL_weekly_return_volatility (1) - AAPL_weekly_return_volatility (1).csv",sep=",")
df
```

```python
df['Label'] = np.where(df['labels']=='Red', 0, 1)
df
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
X_train = df[df['Year']==2020]
X_test = df[df['Year']==2021]
Y_train = X_train['Label']
X_train = X_train.drop(['Label', 'labels', 'Year', 'Week_Number'], axis=1)
Y_test = X_test['Label']
X_test = X_test.drop(['Label', 'labels', 'Year', 'Week_Number'], axis=1)
clf = QuadraticDiscriminantAnalysis()
clf.fit(X_train, Y_train)
Y_pred_QDA = clf.predict(X_test)
print(Y_pred_QDA)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
model = LinearDiscriminantAnalysis()
model.fit(X_train, Y_train)
Y_pred_LDA = model.predict(X_test)
print(Y_pred_LDA)
```

```python
from sklearn.metrics import accuracy_score
QDA_acc = accuracy_score(Y_test, Y_pred_QDA)
LDA_acc = accuracy_score(Y_test, Y_pred_LDA)
print("ACCURACY OF QDA---->", QDA_acc)
print("\n\nACCURACY OF LDA---->", LDA_acc)
```

```python
from sklearn.metrics import confusion_matrix
QDA_conf = confusion_matrix(Y_test, Y_pred_QDA)
LDA_conf = confusion_matrix(Y_test, Y_pred_LDA)
print("CONFUSION MATRIX OF QDA---->\n", QDA_conf)
print("\n\CONFUSION MATRIX OF LDA---->\n", LDA_conf)
```

```python
TPR_QDA = QDA_conf[0][0]/(QDA_conf[0][0]+QDA_conf[1][0])
TNR_QDA = QDA_conf[1][1]/QDA_conf[1][1]+QDA_conf[0][1]
TPR_LDA = LDA_conf[0][0]/(LDA_conf[0][0]+LDA_conf[1][0])
TNR_LDA = LDA_conf[1][1]/LDA_conf[1][1]+LDA_conf[0][1]

print("TPR QDA---->", TPR_QDA)
print("TNR QDA---->", TNR_QDA)
print("TPR LDA---->", TPR_LDA)
print("TNR QDA---->", TNR_LDA)
```
